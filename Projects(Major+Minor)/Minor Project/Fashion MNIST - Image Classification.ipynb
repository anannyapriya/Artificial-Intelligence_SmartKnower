#### Downloading Datasets

```python
(x_train, y_train), (x_test, y_test) = fashion_mnist.load_data()
```

### Preprocessing

```python
x_train = x_train/255
x_test = x_test/255
```

```python
x_train[0]
```

```python
x_train[0].shape
```

```python
x_train.shape
```

#### Taking only first 10,000 datas

```python
x_train = x_train[:10000,:,:]
x_test = x_test[:3000,:,:]
y_train = y_train[:10000]
y_test = y_test[:3000]
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

#### Reshaping the training and test data

```python
x_train = x_train.reshape(10000,28,28,1)
x_test = x_test.reshape(3000,28,28,1)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

### Build the CNN Model

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, Flatten, Dense, MaxPooling2D
```

### Notes
1) Conv2D(no_of_filter, filter_shape, strides, activation, input_shape)<br>
2) MaxPooling2D(pool_size, strides)<br>
3) Flatten()<br>
4) Dense(no_of_neurons, activation)

```python
m1 = Sequential()
m1.add(Conv2D(64, (3,3), strides=1,activation='relu', input_shape=(28,28,1)))
m1.add(MaxPooling2D(pool_size=(2,2), strides=1))
m1.add(Conv2D(64, (3,3), strides=1,activation='relu'))
m1.add(MaxPooling2D(pool_size=(2,2), strides=1))
m1.add(Flatten())
m1.add(Dense(128, activation='relu'))
m1.add(Dense(10, activation='softmax'))

m1.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

#### To run the model

```python
h1 = m1.fit(x_train, y_train, epochs=5, validation_data=(x_test, y_test))
```

#### Printing the following in a table
<b>1) Loss<br>
2) Accuracy<br>
3) Val_Loss<br>
4) Val_Accuracy<br>
5) Epochs</b>

```python
r1 = pd.DataFrame(h1.history)
r1['Epochs'] = h1.epoch
r1.tail()
```

#### Plotting Loss

```python
plt.plot(r1['Epochs'], r1['loss'], label='Training loss')
plt.plot(r1['Epochs'], r1['val_loss'], label='Testing loss')
plt.legend()
plt.show()
```

#### Plotting Accuracy

```python
plt.plot(r1['Epochs'], r1['accuracy'], label='Training accuracy')
plt.plot(r1['Epochs'], r1['val_accuracy'], label='Testing accuracy')
plt.legend()
plt.show()
```

#### Prediction

```python
y_pred = m1.predict(x_test)
print(y_pred)
```

```python
y_pred1 = [np.argmax(i) for i in y_pred]
print(y_pred1)
```

#### To get Confusion Matrix and Classification Matrix, we imported Sklearn Library

```python
from sklearn.metrics import confusion_matrix, classification_report
```

#### Printing the Confusion Matrix and Classification Matrix

```python
print(confusion_matrix(y_test,y_pred1))
print(classification_report(y_test,y_pred1))
```
